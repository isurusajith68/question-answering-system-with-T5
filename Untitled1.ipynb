```python
from transformers import T5Tokenizer, T5ForConditionalGeneration, Trainer, TrainingArguments
from datasets import load_dataset, DatasetDict

# Load pre-trained T5 model and tokenizer
model_name = "t5-small"  # Or "t5-large"
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)

# Load dataset (e.g., SQuAD)
dataset = load_dataset("squad")

# Preprocess dataset
def preprocess_data(examples):
    inputs = ["generate question: " + context for context in examples["context"]]
    targets = [question for question in examples["question"]]
    model_inputs = tokenizer(inputs, max_length=512, truncation=True, padding="max_length")
    labels = tokenizer(targets, max_length=128, truncation=True, padding="max_length").input_ids
    model_inputs["labels"] = labels
    return model_inputs

tokenized_datasets = dataset.map(preprocess_data, batched=True)

# Split dataset
train_size = int(0.5 * len(tokenized_datasets["train"]))  # Use half of the training data
validation_size = int(0.5 * len(tokenized_datasets["validation"]))  # Use half of the validation data

train_dataset = tokenized_datasets["train"].select(range(train_size))
validation_dataset = tokenized_datasets["validation"].select(range(validation_size))

# Training arguments
training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    num_train_epochs=1,
    save_steps=10_000,
    save_total_limit=2,
    logging_dir="./logs"
)

# Define trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=validation_dataset,
    tokenizer=tokenizer,
)

# Train the model
trainer.train()

# Save the model
model.save_pretrained("./t5-qna")
tokenizer.save_pretrained("./t5-qna")
```

```python
import pdfplumber

def extract_text_from_pdf(pdf_path):
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text()
    return text
```

```python
import json

# Define the path to the SQuAD 2.0 training data
path = "squad/train-v2.0.json"

# Load and preprocess the SQuAD 2.0 data
def load_squad_data(path):
    with open(path, 'rb') as f:
        squad_dict = json.load(f)

    texts = []
    questions = []
    answers = []

    for group in squad_dict['data']:
        for passage in group['paragraphs']:
            context = passage['context']
            for qa in passage['qas']:
                question = qa['question']
                for answer in qa['answers']:
                    texts.append(context)
                    questions.append(question)
                    answers.append(answer)

    return texts, questions, answers

# Split the data into a fraction
def split_data(texts, questions, answers, fraction=0.25):
    split_index = int(len(texts) * fraction)  # Determine the split point
    texts_fraction = texts[:split_index]
    questions_fraction = questions[:split_index]
    answers_fraction = answers[:split_index]
    return texts_fraction, questions_fraction, answers_fraction

# Preprocess the data
half_texts, half_queries, half_answers = load_squad_data(path)

# Split the dataset into one-fourth
train_texts, train_queries, train_answers = split_data(half_texts, half_queries, half_answers, fraction=0.25)

# Verify the size of the split data
print(f"Total samples in the dataset: {len(train_texts)}")
print(f"Samples in the quarter dataset: {len(half_texts)}")

# Now `quarter_texts`, `quarter_queries`, and `quarter_answers` contain one-fourth of the original dataset.
# You can proceed to tokenize and train with this subset.
```

```python
def add_token_positions(encodings, answers):
    start_positions = []
    end_positions = []

    count = 0

    for i in range(len(answers)):
        start_positions.append(encodings.char_to_token(i, answers[i]['answer_start']))
        end_positions.append(encodings.char_to_token(i, answers[i]['answer_end']))

        # if start position is None, the answer passage has been truncated
        if start_positions[-1] is None:
            start_positions[-1] = tokenizer.model_max_length

        # if end position is None, the 'char_to_token' function points to the space after the correct token, so add - 1
        if end_positions[-1] is None:
            end_positions[-1] = encodings.char_to_token(i, answers[i]['answer_end'] - 1)

            # if end position is still None the answer passage has been truncated
            if end_positions[-1] is None:
                count += 1
                end_positions[-1] = tokenizer.model_max_length
    print(count)

    # Update the data in dictionary
    encodings.update({'start_positions': start_positions, 'end_positions': end_positions})

add_token_positions(train_encodings, train_answers)
add_token_positions(val_encodings, val_answers)
```

```python
class SquadDataset(torch.utils.data.Dataset):
    def __init__(self, encodings):
        self.encodings = encodings

    def __getitem__(self, idx):
        return {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}

    def __len__(self):
        return len(self.encodings.input_ids)
```

```python
train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=8, shuffle=True)
```

```python
from transformers import AdamW
optim = AdamW(model.parameters(), lr=5e-5)

epochs = 3
```

```python
whole_train_eval_time = time.time()

train_losses = []
val_losses = []

print_every = 1000

for epoch in range(epochs):
    epoch_time = time.time()

    # Set model in train mode
    model.train()
    loss_of_epoch = 0

    print("############Train############")

    for batch_idx,batch in enumerate(train_loader):
        optim.zero_grad()

        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        start_positions = batch['start_positions'].to(device)
        end_positions = batch['end_positions'].to(device)

        outputs = model(input_ids, attention_mask=attention_mask, start_positions=start_positions, end_positions=end_positions)
        loss = outputs[0]
        # do a backwards pass
        loss.backward()
        # update the weights
        optim.step()
        # Find the total loss
        loss_of_epoch += loss.item()

        if (batch_idx+1) % print_every == 0:
            print("Batch {:} / {:}".format(batch_idx+1,len(train_loader)),"\nLoss:", round(loss.item(),1),"\n")

    loss_of_epoch /= len(train_loader)
    train_losses.append(loss_of_epoch)

    ##########Evaluation##################

    # Set model in evaluation mode
    model.eval()

    print("############Evaluate############")

    loss_of_epoch = 0

    for batch_idx,batch in enumerate(val_loader):

        with torch.no_grad():

            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            start_positions = batch['start_positions'].to(device)
            end_positions = batch['end_positions'].to(device)

            outputs = model(input_ids, attention_mask=attention_mask, start_positions=start_positions, end_positions=end_positions)
            loss = outputs[0]
            # Find the total loss
            loss_of_epoch += loss.item()

        if (batch_idx+1) % print_every == 0:
            print("Batch {:} / {:}".format(batch_idx+1,len(val_loader)),"\nLoss:", round(loss.item(),1),"\n")

    loss_of_epoch /= len(val_loader)
    val_losses.append(loss_of_epoch)

    # Print each epoch's time and train/val loss

    print("\n-------Epoch ", epoch+1,
          "-------"
          "\nTraining Loss:", train_losses[-1],
          "\nValidation Loss:", val_losses[-1],
          "\nTime: ",(time.time() - epoch_time),
          "\n-----------------------",
          "\n\n")

print("Total training and evaluation time: ", (time.time() - whole_train_eval_time))
```

```python
def process_pdf_and_evaluate_answers(pdf_path, model, tokenizer, max_context_length=256):
    # Extract text from the PDF
    text = extract_text_from_pdf(pdf_path)

    # Check if text extraction is successful
    if not text:
        raise ValueError("The extracted text from the PDF is empty!")

    # Split the text into sentences or paragraphs for better context
    sentences = text.split('. ')  # Split by period-space to get sentences, you can adjust this if needed

    # Rebuild the chunks, ensuring we don't break in the middle of sentences
    chunks = []
    current_chunk = ""

    for sentence in sentences:
        if len(current_chunk) + len(sentence) < max_context_length:
            # Add sentence to the current chunk
            current_chunk += sentence + ". "
        else:
            # If adding the sentence exceeds the max length, save the current chunk
            chunks.append(current_chunk.strip())
            current_chunk = sentence + ". "  # Start a new chunk with the current sentence

    # Add any remaining text as the last chunk
    if current_chunk:
        chunks.append(current_chunk.strip())

    # Store QA pairs and evaluation results
    evaluation_results = []

    for chunk in chunks:
        # Check if chunk is non-empty
        if len(chunk.strip()) == 0:
            continue  # Skip empty or invalid chunks

        try:
            # Generate a question based on the context
            question = generate_questions(chunk, model, tokenizer)

            # If the question is empty, skip this chunk
            if not question.strip():
                continue

            # Predicted answer (empty for initial testing or provided later)
            true_answer = ""  # Modify with actual answers if available
            prediction = predict(chunk, question)

            # Evaluate the prediction
            em_score = compute_exact_match(prediction, true_answer)
            f1_score = compute_f1(prediction, true_answer)

            # Store the results
            evaluation_results.append({
                "context": chunk,
                "question": question,
                "true_answer": true_answer,
                "prediction": prediction,
                "em_score": em_score,
                "f1_score": f1_score,
            })

        except Exception as e:
            print(f"Error processing chunk: {chunk[:50]}... Error: {str(e)}")
            continue  # Skip problematic chunk and continue with the next

    return evaluation_results

# Path to your PDF file
pdf_path = "technology_paragraphs.pdf"

# Process the PDF and evaluate answers
results = process_pdf_and_evaluate_answers(pdf_path, model, tokenizer)

# Print results
for i, result in enumerate(results, 1):
    print(f"Question {i}: {result['question']}")
    print(f"Prediction: {result['prediction']}")
    print(f"True Answer: {result['true_answer']}")
    print(f"Context: {result['context']}")
    print(f"EM Score: {result['em_score']}")
    print(f"F1 Score: {result['f1_score']}\n")
```

```python
import torch
from transformers import T5ForConditionalGeneration, T5Tokenizer
import string
import re
from PyPDF2 import PdfReader

# Load the fine-tuned T5 model and tokenizer
model = T5ForConditionalGeneration.from_pretrained("./t5-qna")
tokenizer = T5Tokenizer.from_pretrained("./t5-qna")

# Function to extract text from a PDF file
def extract_text_from_pdf(pdf_path):
    pdf_reader = PdfReader(pdf_path)
    text = ""
    for page in pdf_reader.pages:
        text += page.extract_text()
    return text

# Function to generate a question from a chunk of text using T5
def generate_questions(chunk, model, tokenizer):
    input_text = f"generate question: {chunk}"
    
    # Tokenize the input
    inputs = tokenizer(input_text, return_tensors='pt', padding=True, truncation=True)

    # Generate the output (question)
    outputs = model.generate(inputs['input_ids'], max_length=150, num_beams=5, early_stopping=True, temperature=0.6)

    # Decode the generated question
    question = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return question

# Function to generate an answer using T5
def predict(context, query):
    input_text = f"question: {query} context: {context}"
    
    # Tokenize the input
    inputs = tokenizer(input_text, return_tensors='pt', padding=True, truncation=True)
    
    # Generate the output (answer) with adjusted parameters
    outputs = model.generate(inputs['input_ids'], max_length=150, num_beams=8, early_stopping=True, temperature=0.5)
    
    # Decode the generated answer and clean it up
    answer = tokenizer.decode(outputs[0], skip_special_tokens=True)
    
    # Clean up repetitive answers (basic filtering example)
    if answer == query:
        answer = "The answer is unclear, please try again."
    
    return answer

# Function to normalize text for comparison (used in EM and F1 calculations)
def normalize_text(s):
    def remove_articles(text):
        regex = re.compile(r"\b(a|an|the)\b", re.UNICODE)
        return re.sub(regex, " ", text)

    def white_space_fix(text):
        return " ".join(text.split())

    def remove_punc(text):
        exclude = set(string.punctuation)
        return "".join(ch for ch in text if ch not in exclude)

    def lower(text):
        return text.lower()

    return white_space_fix(remove_articles(remove_punc(lower(s))))

# Compute Exact Match (EM) score
def compute_exact_match(prediction, truth):
    return int(normalize_text(prediction) == normalize_text(truth))

# Compute F1 score
def compute_f1(prediction, truth):
    pred_tokens = normalize_text(prediction).split()
    truth_tokens = normalize_text(truth).split()

    if len(pred_tokens) == 0 or len(truth_tokens) == 0:
        return int(pred_tokens == truth_tokens)

    common_tokens = set(pred_tokens) & set(truth_tokens)

    if len(common_tokens) == 0:
        return 0

    prec = len(common_tokens) / len(pred_tokens)
    rec = len(common_tokens) / len(truth_tokens)
    return 2 * (prec * rec) / (prec + rec)

# Main function to process the PDF, generate questions and predict answers
def process_pdf_and_generate_questions_with_context(pdf_path, model, tokenizer, max_context_length=512):
    # Extract text from the PDF
    text = extract_text_from_pdf(pdf_path)

    # Split text into manageable chunks based on context size
    chunks = [text[i:i + max_context_length] for i in range(0, len(text), max_context_length)]

    qa_pairs = []
    for chunk in chunks:
        # Generate a question from the chunk
        question = generate_questions(chunk, model, tokenizer)
        
        # Generate an answer using the model
        answer = predict(chunk, question)

        # Store the context, question, and answer
        qa_pairs.append({"context": chunk, "question": question, "answer": answer})

    return qa_pairs

# Example usage
pdf_path = "technology_paragraphs_removed.pdf"

# Generate questions and answers with context from the PDF
qa_pairs = process_pdf_and_generate_questions_with_context(pdf_path, model, tokenizer)

# Print generated questions with context and predicted answers
for i, qa in enumerate(qa_pairs, 1):
    print(f"Question {i}: {qa['question']}")
    print(f"Context: {qa['context']}")
    print(f"Predicted Answer: {qa['answer']}\n")
```
